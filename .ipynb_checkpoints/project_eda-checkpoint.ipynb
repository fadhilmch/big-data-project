```python
import pandas as pd
import numpy as np
from wordcloud import WordCloud
import matplotlib.pyplot as plt
from sklearn.feature_extraction.text import CountVectorizer
from scipy.stats import hmean,norm
from bokeh.plotting import figure
from bokeh.io import output_notebook, show
from bokeh.models import LinearColorMapper
from bokeh.models import HoverTool
```

```python
df_cleaned_dataset = pd.read_csv('data/cleaned_tweets.csv')
df_cleaned_dataset.head()
```

### Word Cloud

```python
def plot_wordcloud(data, colormap='viridis'):
    words = []
    for tweet in data.tweet:
        words.append(tweet)
        
    words = pd.Series(words).str.cat(sep=' ')
    
    wordcloud = WordCloud(width=1000, height=500, max_font_size=150, colormap=colormap).generate(words)
    plt.figure(figsize=(10,8))
    plt.imshow(wordcloud, interpolation="bilinear")
    plt.axis("off")
    plt.show
```

```python
# wordcloud negative
plot_wordcloud(df_cleaned_dataset[df_cleaned_dataset.target==0], 'magma')
```

```python
# wordcloud positive
plot_wordcloud(df_cleaned_dataset[df_cleaned_dataset.target==1])
```

### Word Frequency

```python
def get_word_freq(data, max_features, stop_words=False):
    if stop_words:
        count_vec = CountVectorizer(stop_words='english', max_features=max_features)
    else:
        count_vec = CountVectorizer(max_features=max_features)
        
    count_vec.fit(data.tweet)
    print("number of unique words : " + str(len(count_vec.get_feature_names())))
    
    matrix_neg = count_vec.transform(data[data.target == 0].tweet)
    matrix_pos = count_vec.transform(data[data.target == 1].tweet)
    freq_neg = np.sum(matrix_neg, axis=0)
    freq_pos = np.sum(matrix_pos, axis=0)
    neg = np.squeeze(np.asarray(freq_neg))
    pos = np.squeeze(np.asarray(freq_pos))
    df_freq = pd.DataFrame([neg,pos],columns=count_vec.get_feature_names()).transpose()
    df_freq.columns = ['neg', 'pos']
    df_freq['total'] = df_freq['neg'] + df_freq['pos']
    df_freq = df_freq.sort_values('total', ascending=False)
    
    return df_freq
```

```python
def plot_top_words(data, n, sentiment):
    y = np.arange(n)
    plt.figure(figsize=(10,8))
    plt.bar(y, data.sort_values(by=sentiment, ascending=False)[sentiment][:n], align='center', alpha=0.5)
    plt.xticks(y, data.sort_values(by=sentiment, ascending=False).index[:n], rotation=90)
    plt.xlabel('word')
    plt.ylabel('freq')
```

```python
count_vec = CountVectorizer()
count_vec.fit(df_cleaned_dataset.tweet)
print("number of unique words : " + str(len(count_vec.get_feature_names())))
```

```python
matrix_neg = count_vec.transform(df_cleaned_dataset[df_cleaned_dataset.target == 0].tweet)
matrix_pos = count_vec.transform(df_cleaned_dataset[df_cleaned_dataset.target == 1].tweet)
freq_neg = np.sum(matrix_neg, axis=0)
freq_pos = np.sum(matrix_pos, axis=0)
neg = np.squeeze(np.asarray(freq_neg))
pos = np.squeeze(np.asarray(freq_pos))
df_freq = pd.DataFrame([neg,pos],columns=count_vec.get_feature_names()).transpose()
df_freq.columns = ['neg', 'pos']
df_freq['total'] = df_freq['neg'] + df_freq['pos']
df_freq = df_freq.sort_values('total', ascending=False)
df_freq.head()
```

```python
df_freq.to_csv('data/word_freq.csv')
```

```python
def normcdf(data):
    return norm.cdf(data, data.mean(), data.std())
```

```python
df_freq_2 = df_freq.copy()
df_freq_2['pos_rate'] = df_freq_2['pos']/df_freq_2['total']
df_freq_2['pos_pct'] = df_freq_2['pos']/df_freq_2['pos'].sum()
df_freq_2['pos_hmean'] = df_freq_2.apply(lambda x: (hmean([x.pos_rate, x.pos_pct]) if x.pos_rate > 0 and x.pos_pct > 0 else 0), axis=1)
df_freq_2['pos_rate_cdf'] = normcdf(df_freq_2['pos_rate'])
df_freq_2['pos_pct_cdf'] = normcdf(df_freq_2['pos_pct'])
df_freq_2['pos_cdf_hmean'] = hmean([df_freq_2.pos_rate_cdf, df_freq_2.pos_pct_cdf])
#df_freq_2.sort_values(by='pos_rate', ascending=False).head(10)
#df_freq_2.sort_values(by='pos_pct', ascending=False).head(10)
#df_freq_2.sort_values(by='pos_hmean', ascending=False).head(10)
df_freq_2.sort_values(by='pos_cdf_hmean', ascending=False).head(10)
```

```python
def get_word_importance(data):
    # positive word importance
    data['pos_rate'] = data['pos']/data['total']
    data['pos_pct'] = data['pos']/data['pos'].sum()
    data['pos_hmean'] = data.apply(lambda x: (hmean([x.pos_rate, x.pos_pct]) if x.pos_rate > 0 and x.pos_pct > 0 else 0), axis=1)
    data['pos_rate_cdf'] = normcdf(data['pos_rate'])
    data['pos_pct_cdf'] = normcdf(data['pos_pct'])
    data['pos_cdf_hmean'] = hmean([data.pos_rate_cdf, data.pos_pct_cdf])
    
    # negative word importance
    data['neg_rate'] = data['neg']/data['total']
    data['neg_pct'] = data['neg']/data['neg'].sum()
    data['neg_hmean'] = data.apply(lambda x: (hmean([x.neg_rate, x.neg_pct]) if x.neg_rate > 0 and x.neg_pct > 0 else 0), axis=1)
    data['neg_rate_cdf'] = normcdf(data['neg_rate'])
    data['neg_pct_cdf'] = normcdf(data['neg_pct'])
    data['neg_cdf_hmean'] = hmean([data.neg_rate_cdf, data.neg_pct_cdf])
    
    return data[['neg', 'pos', 'total', 'neg_cdf_hmean', 'pos_cdf_hmean']]
```

```python
df_freq_2 = get_word_importance(df_freq)
```

```python
df_freq_2.sort_values(by='neg_cdf_hmean', ascending=False).head(10)
```

```python
df_freq_2.sort_values(by='pos_cdf_hmean', ascending=False).head(10)
```

```python
output_notebook()
color_mapper = LinearColorMapper(palette='Viridis256', low=min(df_freq_2.pos_cdf_hmean), high=max(df_freq_2.pos_cdf_hmean))
p = figure(x_axis_label='neg_cdf_hmean', y_axis_label='pos_cdf_hmean')
p.circle('neg_cdf_hmean','pos_cdf_hmean',size=5,alpha=0.3,source=df_freq_2,color={'field': 'pos_cdf_hmean', 'transform': color_mapper})
hover = HoverTool(tooltips=[('token','@index')])
p.add_tools(hover)
show(p)
```
